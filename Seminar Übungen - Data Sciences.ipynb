### Das `pandas`-Paket einbinden

```python
import pandas as pd
```

### csv-Datei mit `read_ csv()` einlesen, dazu das Komma als delimiter spezifizieren `delimiter=","`

```python
patienten_df = pd.read_csv("https://raw.githubusercontent.com/gaetanwabo/synthea_data/main/encounters.csv", delimiter=",")
```

### Sich einen ersten Überblick über die Daten verschaffen:  `head()` -Methode zur Ausgabe des DataFrame (Tabelle)

```python
patienten_df.head()
```

###  `len()`-Methode: Ausgabe der Anzahl der  Tabelle-Einträge

```python
len(patienten_df)
```

###  `columns.str.lower()`-Methode : Spaltenamen kleinschreiben

```python
patienten_df.columns = patienten_df.columns.str.lower()
patienten_df
```

###   `isnull().sum()` -Methode: Überprüfung von fehlenden Werten (Datenqualität)

```python
patienten_df.isnull().sum()
```

### Auf eine Spalte eines `DataFrames` zugreifen

Mit der `[ ] - Schreibweise` können wir über den Titel auf die Inhalte einer Spalte zugreifen - so wie bei einem dictionary über den Schlüssel.

```python
patienten_df['patient']
```

### `loc()` verwenden, um mehrere Spalten  mit Namen anzuzeigen: 

```python
patienten_df.loc[:,['patient','reasoncode']]
```

### Auf eine Zeile eines `DataFrames` zugreifen

Für den Zugriff auf eine bestimmte Spalte brauchen wir die **`iloc()`**-Methode (die `[ ] - Schreibweise` ist ja schon mit dem Zugriff auf die Spalten belegt).

```python
patienten_df.iloc[100]
```

### Einfache bzw. Mehrfache Filter
Natürlich kann man einfache oder mehrere Filter an ein `DataFrame` anlegen. 
Achte darauf, wenn du in mehrfachen Stufen filtern möchtest, gefilterte `DataFrames` in neuen Variablen zwischenzuspeichern: 

```python
#z.B. man will Patienten auswerten, die 
#1. ambulant versorgt werden und,
#2. wegen eines Besuch vor der Geburt in die Praxis kommen, und,
#3. bei denen die Gesamtversorgungskosten großer 100 Dollar sind


filter1 = patienten_df[patienten_df["encounterclass"] == "ambulatory"] ## Patienten aus dem Ambulanten Bereich
filter2 = filter1[filter1["description"] == "Prenatal visit"]  
filter3 = filter2[filter2["base_encounter_cost"] > 100]
end_cohort= filter3
end_cohort
```
